```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np
import plotly.express as px
```

```python
df = pd.read_csv("LaLiga newdataset.csv")
df.drop(columns=['Unnamed: 0','round'],inplace=True)
df
```

```python
df.isnull().sum()
df['captain'].unique().tolist()
df[df['captain'].isnull() == True].dropna()
# df['captain'].fillna()
mode_A = df['captain'].mode()[0]
# df['captain'] = df['captain'].fillna(mode_A)
df.dropna(subset=['captain'], inplace=True)
df.isnull().sum()
```

```python
df[df['attendance'].isnull() == True]
# df.dropna(subset=['attendance'],inplace=True)
attendence = df['attendance'].mode()[0]
attendence
df['attendance'] = df['attendance'].fillna(attendence)
df.isnull().sum()
```

```python
poss = df['poss'].mode()[0]
poss
df['poss'] = df['poss'].fillna(attendence)
df.isnull().sum()
```

```python
df['g/sot'].unique().tolist()
poss = df['g/sot'].mode()[0]
poss
df['g/sot'] = df['g/sot'].fillna(attendence)
df.isnull().sum()
```
